## Making an EZ WordCloud of the Disney Cars Movie Script


#### Hiii  just for funnnnn lets make a WordCloud of The Disney Cars Movie! 

I'm using the cars movie script <a href = "https://www.nationstates.net/page=dispatch/id=768849"> from this website </a>  

First, save the text to a txt file named 'wc.txt' and let's open it:

```python

text = ""
with open('wc.txt', encoding='utf-8') as f:
    text = ''.join(f.readlines())
```

Then, we're gonna need the wordcloud package for python:
`pip install wordcloud`

Import the necessary packages: 
    

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
%matplotlib inline

from PIL import Image
import numpy as np
```

Make a WordCloud Object and a Custom Mask: (I'm using <a href = "https://www.silhouette.pics/172920/lightning-mcqueen-car-silhouette-vector.php"> this silhouette image </a> of McQueen)

A custom mask lets you shape your wordcloud to the mask, instead of a boring old square. 

```python
    
wc = WordCloud()

mask = np.array(Image.open("lightning-mcqueen-car-silhouette-vector.jpg"))
wc = WordCloud(background_color="white", mask=mask)
wc.generate(text)


plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

TADAA, all done!
You can save your image to a file too: 

```python
wc.to_file("cars-wc.png")
```
